```python
import pandas as pd
from scipy.stats import pearsonr

# Load CSV file into DataFrame
df = pd.read_csv("/mnt/HDD3/khanh/tplam/MOICVAE/data/Datasets/Drug response plam/gdsc_gene_expression.csv")  # Replace with your actual file name
```

```python
# Set the first column (e.g., sample name or ID) as the index
df.set_index(df.columns[0], inplace=True)

# Ensure all data is numeric
df_numeric = df.apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values (optional)
df_clean = df_numeric.dropna()

# Prepare an empty DataFrame to store row-wise correlations
row_names = df_clean.index
corr_matrix = pd.DataFrame(index=row_names, columns=row_names, dtype=float)

# Compute pairwise Pearson correlation between rows
for i in row_names:
    for j in row_names:
        if pd.isna(corr_matrix.loc[i, j]):  # Avoid duplicate calculations
            corr, _ = pearsonr(df_clean.loc[i], df_clean.loc[j])
            corr_matrix.loc[i, j] = corr
            corr_matrix.loc[j, i] = corr  # Mirror it since correlation is symmetric

print("Row-wise Pearson Correlation Matrix:\n")
print(corr_matrix)
```

```python
df_T = df.T
# df_T.set_index(df_T.columns[0], inplace=True)

# Convert all remaining columns to numeric if they aren't already
df_numeric = df_T.apply(pd.to_numeric, errors='coerce')

# Drop rows with any missing values (optional, depending on your needs)
df_clean = df_numeric.dropna()

# Option 1: Pearson correlation matrix for all numeric columns
correlation_matrix = df_clean.corr(method='pearson')
correlation_matrix.to_csv("./data/Datasets/Drug response plam/correlation_matrix.csv")
```

```python
import pandas as pd

# Load CSV
df = pd.read_csv("data/Datasets/Drug response plam/gdsc_log_ic50.csv")  # Replace with your file name

# Separate ID column
ids = df.iloc[:, 0]
data = df.iloc[:, 1:]

# Convert to numeric in case of any string issues
data = data.apply(pd.to_numeric, errors='coerce')

# Compute medians for each column
medians = data.median()

# Apply transformation: 1 if <= median, else 0
binary_data = data.le(medians).astype(int)

# Combine ID column back in
output_df = pd.concat([ids, binary_data], axis=1)

# Save to new CSV
output_df.to_csv("./data/Datasets/Drug response plam/binary_matrix.csv", index=False)

print("✅ Output saved to 'output.csv'")
```
